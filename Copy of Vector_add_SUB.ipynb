```python
import numpy as np

x = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])
weights_input_hidden = np.random.rand(2, 2)
weights_hidden_output = np.random.rand(2, 1)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

hidden_layer = sigmoid(np.dot(x, weights_input_hidden))
output_layer = sigmoid(np.dot(hidden_layer, weights_hidden_output))
error = y - output_layer
d_predicted_output = error * sigmoid(output_layer)
error_hidden_layer = d_predicted_output.dot(weights_hidden_output.T)
d_hidden_layer = error_hidden_layer * sigmoid(hidden_layer)
weights_hidden_output += hidden_layer.T.dot(d_predicted_output)
weights_input_hidden += x.T.dot(d_hidden_layer)

epochs = 10000
print("\nOutput Layer after {} epochs:".format(epochs))
print(output_layer)
```

```python
print("Error :", error)
```
